```python
from collections import Counter, defaultdict

import pandas as pd
import matplotlib.pyplot as plt
import logomaker


def _make_probability_matrix(sequences):
    """
    Build position-wise probability matrix for equal-length sequences.
    """
    if not sequences:
        raise ValueError("Empty sequence list.")

    seq_len = len(sequences[0])
    if any(len(s) != seq_len for s in sequences):
        raise ValueError("All sequences must have the same length inside one group.")

    alphabet = sorted(set("".join(sequences)))
    matrix = []

    for i in range(seq_len):
        col = [seq[i] for seq in sequences]
        counts = Counter(col)
        total = sum(counts.values())
        matrix.append({aa: counts.get(aa, 0) / total for aa in alphabet})

    return pd.DataFrame(matrix)

AA_COLOR_SCHEME = {
    'A': '#1f77b4',
    'C': '#ff7f0e',
    'D': '#d62728',
    'E': '#d62728',
    'F': '#2ca02c',
    'G': '#17becf',
    'H': '#9467bd',
    'I': '#2ca02c',
    'K': '#1f77b4',
    'L': '#2ca02c',
    'M': '#2ca02c',
    'N': '#9467bd',
    'P': '#8c564b',
    'Q': '#9467bd',
    'R': '#1f77b4',
    'S': '#e377c2',
    'T': '#e377c2',
    'V': '#2ca02c',
    'W': '#bcbd22',
    'Y': '#bcbd22',
}


def _collect_weblogo_panels(sequences, min_count=1):
    """Collect one logo matrix per sequence length."""
    seqs = [str(s).strip().upper() for s in sequences if str(s).strip()]
    if not seqs:
        return []

    groups = defaultdict(list)
    for s in seqs:
        groups[len(s)].append(s)

    panels = []
    for length, vals in sorted(groups.items()):
        if len(vals) < min_count:
            continue
        panels.append((length, _make_probability_matrix(vals), len(vals)))

    return panels


def _collect_method_rows(df, min_count=1):
    """Flatten cluster logos into a vertical list of rows for one method."""
    rows = []
    for cid, group in df.groupby('cid', sort=False):
        panels = _collect_weblogo_panels(group['cdr3aa'].tolist(), min_count=min_count)
        for panel_idx, panel in enumerate(panels):
            length, pwm, n = panel
            rows.append({
                'cid': cid,
                'length': length,
                'pwm': pwm,
                'n': n,
                'show_cid': panel_idx == 0,
            })
    return rows


def _plot_logo_on_axis(ax, row, column_title=None):
    """Render one weblogo row with explicit amino-acid coloring."""
    if row is None:
        ax.axis('off')
        return

    color_scheme = {aa: AA_COLOR_SCHEME.get(aa, '#7f7f7f') for aa in row['pwm'].columns}
    logomaker.Logo(row['pwm'], ax=ax, color_scheme=color_scheme)
    ax.set_ylabel('Probability')
    ax.set_xlabel('Position')

    header = f"{column_title}\n{row['cid']}" if column_title else (row['cid'] if row['show_cid'] else '')
    if header:
        ax.set_title(f"{header}\nlen={row['length']}, n={row['n']}")
    else:
        ax.set_title(f"len={row['length']}, n={row['n']}")


def plot_weblogo_length_panels(
    redcea_df,
    tcrnet_df,
    redcea_title='RedCEA',
    tcrnet_title='TCRnet',
    title=None,
    min_count=1,
):
    """Plot one figure with all RedCEA clusters on the left and all TCRnet clusters on the right."""
    redcea_rows = _collect_method_rows(redcea_df, min_count=min_count)
    tcrnet_rows = _collect_method_rows(tcrnet_df, min_count=min_count)

    if not redcea_rows and not tcrnet_rows:
        raise ValueError('No valid sequences to plot.')

    nrows = max(len(redcea_rows), len(tcrnet_rows))
    fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(16, 2.8 * nrows), squeeze=False)

    for row_idx in range(nrows):
        left_row = redcea_rows[row_idx] if row_idx < len(redcea_rows) else None
        right_row = tcrnet_rows[row_idx] if row_idx < len(tcrnet_rows) else None
        _plot_logo_on_axis(axes[row_idx, 0], left_row, redcea_title if row_idx == 0 else None)
        _plot_logo_on_axis(axes[row_idx, 1], right_row, tcrnet_title if row_idx == 0 else None)

    if title:
        fig.suptitle(title, y=1.0)
        fig.tight_layout(rect=(0, 0, 1, 0.99))
    else:
        fig.tight_layout()
    return fig, axes, {'redcea': redcea_rows, 'tcrnet': tcrnet_rows}
```

```python
redcea_plot_df = ylq_redcea_csar.copy()
tcrnet_plot_df = ylq_tcrnet_csar.copy()

redcea_plot_df['cid'] = pd.Categorical(
    redcea_plot_df['cid'],
    categories=redcea_plot_df['cid'].value_counts().index.tolist(),
    ordered=True,
)
tcrnet_plot_df['cid'] = pd.Categorical(
    tcrnet_plot_df['cid'],
    categories=tcrnet_plot_df['cid'].value_counts().index.tolist(),
    ordered=True,
)

redcea_plot_df = redcea_plot_df.sort_values('cid')
tcrnet_plot_df = tcrnet_plot_df.sort_values('cid')
```

```python
plot_weblogo_length_panels(
    redcea_plot_df,
    tcrnet_plot_df,
    redcea_title='RedCEA',
    tcrnet_title='TCRnet',
    title='YLQPRTFLL / CSAR cluster comparison',
    min_count=1,
)
plt.show()
```
